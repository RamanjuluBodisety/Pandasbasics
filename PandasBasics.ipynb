### 5 What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

```python
import pandas as pd

# Creating a DataFrame
sales_data = {
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 'Furniture', 'Electronics', 'Electronics', 'Furniture'],
    'Sales': [200, 150, 300, 400, 120, 500, 100, 220]
}
df = pd.DataFrame(sales_data)
print("Original DataFrame:")
print(df)
```

```python
# Grouping by 'Category' and summing the 'Sales'
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

print("\nTotal Sales by Category:")
print(category_sales)
```

#### Example of Mutability:
Series:

```python
import pandas as pd

# Creating a Series
s = pd.Series([1, 2, 3, 4, 5])

# Modifying the first element
s[0] = 10
print("Modified Series:")
print(s)
```

```python
import pandas as pd

# Creating a DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Modifying a value in the DataFrame
df.at[0, 'A'] = 10
print("Modified DataFrame:")
print(df)
```

#### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Creating individual Series
name_series = pd.Series(['Alice', 'Bob', 'Charlie', 'David', 'Eva'])
age_series = pd.Series([24, 30, 22, 35, 28])
gender_series = pd.Series(['Female', 'Male', 'Male', 'Male', 'Female'])

# Creating a DataFrame using the Series
data = {
    'Name': name_series,
    'Age': age_series,
    'Gender': gender_series
}

df = pd.DataFrame(data)

# Printing the DataFrame
print(df)
```
